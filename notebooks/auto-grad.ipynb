```python
from minitorch.scalar import Scalar
from minitorch.tensor import tensor
```

```python
def f(a, b, c):
    return a + b + c
```

```python
a, b, c = Scalar(1), Scalar(2), Scalar(3)
out = f(a, b, c)
out.backward()
a.derivative, b.derivative, c.derivative
```

```python
tensor1 = tensor([1, 2, 3])
out = tensor1.sum()
out.backward()
# shape (3,)
tensor1.grad
```
